```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 0)
```

```python
# é‡å›žå¸°ã®ç·šå½¢å›žå¸°ã‚’å®Ÿè¡Œ
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
# é‡å›žå¸°ç·šå½¢ãƒ¢ãƒ‡ãƒ«ã®å­¦ç¿’
# trainãƒ‡ãƒ¼ã‚¿ã®ã¿ã‚’ç”¨ã„ã¦ã„ã‚‹ã“ã¨ã«æ³¨æ„
lr.fit(X_train, y_train)
```

```python
# æ±ºå®šä¿‚æ•°ã‚’trainã¨testã«åˆ†ã‘ã¦å‡ºåŠ›
# éŽå­¦ç¿’ã—ã¦ã„ã‚‹ã‹ã©ã†ã‹ã‚’ç¢ºèª
print('R^2')
print('train: %.3f' % lr.score(X_train, y_train))
print('test : %.3f' % lr.score(X_test, y_test))
```

```python
# è‡ªç”±åº¦èª¿æ•´æ¸ˆã¿æ±ºå®šä¿‚æ•°
# (æ±ºå®šä¿‚æ•°, trainã¾ãŸã¯testã®ã‚µãƒ³ãƒ—ãƒ«æ•°, åˆ©ç”¨ã—ãŸç‰¹å¾´é‡ã®æ•°)
def adjusted(score, n_sample, n_features):
    adjusted_score = 1 - (1 - score) * ((n_sample - 1) / (n_sample - n_features - 1))
    return adjusted_score
# è‡ªç”±åº¦èª¿æ•´æ¸ˆã¿æ±ºå®šä¿‚æ•°ã‚’å‡ºåŠ›
print('adjusted R^2')
print('train: %3f' % adjusted(lr.score(X_train, y_train), len(y_train), 2))
print('test : %3f' % adjusted(lr.score(X_test, y_test), len(y_test), 2))
```

```python
# MSEã‚’å‡ºåŠ›ã™ã‚‹é–¢æ•°ã‚’èª­ã¿è¾¼ã‚€
from sklearn.metrics import mean_squared_error as mse

# RMSEã‚’trainã¨testã«åˆ†ã‘ã¦å‡ºåŠ›
# éŽå­¦ç¿’ã‚’ã—ã¦ã„ã‚‹ã‹ã©ã†ã‹ã‚’ç¢ºèª
print('RMSE')
print('train: %.3f' % (mse(y_train, lr.predict(X_train)) ** (1/2)))
print('test : %.3f' % (mse(y_test, lr.predict(X_test)) ** (1/2)))
```

```python
# æ®‹å·®ãƒ—ãƒ­ãƒƒãƒˆã‚’è¡Œã†é–¢æ•°ã‚’å®šç¾©
def res_plot(y_train, y_train_pred, y_test, y_test_pred):
    # train, testã”ã¨ã«æ®‹å·®ã‚’ç®—å‡º
    res_train = y_train_pred - y_train
    res_test  = y_test_pred - y_test
    
    # train, testã”ã¨ã«(äºˆæ¸¬å€¤, æ®‹å·®)ã‚’ãƒ—ãƒ­ãƒƒãƒˆ
    plt.figure(figsize=(8, 8))
    plt.scatter(y_train_pred, res_train, color='blue', marker='o', label='train', alpha=0.5)
    plt.scatter(y_test_pred, res_test, color='green', marker='s', label='test', alpha=0.5)
    
    plt.xlabel('Predicted Values')                  # xè»¸ã®ãƒ©ãƒ™ãƒ«ã‚’è¡¨ç¤º
    plt.ylabel('Residuals')                         # yè»¸ã®ãƒ©ãƒ™ãƒ«ã‚’è¡¨ç¤º
    plt.legend(loc='upper left')                    # å‡¡ä¾‹ã‚’å·¦ä¸Šã«è¡¨ç¤º
    plt.hlines(y=0, xmin=-10, xmax=50, color='red') # æ®‹å·®0ã®ç·šã‚’è¡¨ç¤º
    plt.xlim([-10, 50])                             # è¡¨ç¤ºã™ã‚‹æ¨ªè»¸ã®ç¯„å›²ã‚’æŒ‡å®š
    plt.show()
# æ®‹å·®ãƒ—ãƒ­ãƒƒãƒˆã‚’å‡ºåŠ›
res_plot(y_train, lr.predict(X_train), y_test, lr.predict(X_test))
```

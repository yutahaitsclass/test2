```python
# 必要なライブラリの読み込み
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
% matplotlib inline
import seaborn as sns

import seaborn as sns
from sklearn.preprocessing import StandardScaler
from sklearn.svm import SVC
from sklearn import metrics
from sklearn.model_selection import StratifiedKFold
from sklearn.model_selection import GridSearchCV
```

```python
# trainデータ
train = pd.read_csv('train.csv')
```

```python
# データ数の確認
print(train.shape)
```

```python
# 変数の確認
# len(train.columns)
train.columns
```

```python
# 欠損値の確認
train.isnull().sum()[train.isnull().sum()>0].sort_values(ascending=False)
```

```python
train = train.drop(columns=["PoolQC","MiscFeature","Alley","Fence" ,"FireplaceQu" ,"LotFrontage","LotFrontage"])
```

```python
# 平均値補完
train=train.fillna(train.mean())
```

```python
print(train.dtypes)
```

```python
# string label to categorical values
from sklearn.preprocessing import LabelEncoder

for i in range(train.shape[1]):
    if train.iloc[:,i].dtypes == object:
        lbl = LabelEncoder()
        lbl.fit(list(train.iloc[:,i].values))
        train.iloc[:,i] = lbl.transform(list(train.iloc[:,i].values))
     
```

```python
sns.distplot(train["SalePrice"])
plt.show()
```

```python
np.abs(train.corr()["SalePrice"]).sort_values()
```

```python
X = train.loc[:,train.columns != "SalePrice"]
y= train.loc[:,["SalePrice"]]
```

```python
# データセットの分割
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=0)
```

```python
from sklearn.ensemble import RandomForestClassifier
rfc = RandomForestClassifier(random_state=0, n_estimators=100)
rfc.fit(X_train, y_train)
```

```python
# 特徴量の重要度をグラフに出力
n_features = len(X.columns)
plt.figure(figsize=(34, 26))
plt.barh(range(n_features), rfc.feature_importances_ , align='center')
plt.yticks(np.arange(n_features), X.columns)
plt.show()
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X_pickup, y, test_size = 0.3, random_state = 0)
```

```python
from sklearn.linear_model import LinearRegression

lr = LinearRegression()

lr.fit(X_train, y_train)
```

```python
print(lr.score(X_test, y_test))
```

```python
# 自由度調整済み決定係数
# (決定係数, trainまたはtestのサンプル数, 利用した特徴量の数)
def adjusted(score, n_sample, n_features):
    adjusted_score = 1 - (1 - score) * ((n_sample - 1) / (n_sample - n_features - 1))
    return adjusted_score
```

```python
# 線形回帰
print('model_linear')
print('train: %.3f' % adjusted(lr.score(X_train, y_train), len(y_train), 1))
print('test : %.3f' % adjusted(lr.score(X_test, y_test), len(y_test), 1))
print('')
```
